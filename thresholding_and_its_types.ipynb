**All Thresholding Approaches**

```python
# ==========================
# All Thresholding Approaches
# ==========================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# --------------------------
# Step 1: Upload an Image
# --------------------------
uploaded = files.upload()
for filename in uploaded.keys():
    img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)  # convert to grayscale

# --------------------------
# Step 2: Global Thresholding
# --------------------------
# 1. Binary
_, th_binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
# 2. Binary Inverse
_, th_binary_inv = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
# 3. Truncate
_, th_trunc = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
# 4. ToZero
_, th_tozero = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
# 5. ToZero Inverse
_, th_tozero_inv = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)

# --------------------------
# Step 3: Adaptive Thresholding
# --------------------------
# 1. Adaptive Mean
th_adaptive_mean = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                         cv2.THRESH_BINARY, 11, 2)
# 2. Adaptive Gaussian
th_adaptive_gaussian = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                             cv2.THRESH_BINARY, 11, 2)

# --------------------------
# Step 4: Otsu's Thresholding
# --------------------------
# Apply Gaussian blur first (optional but recommended)
blur = cv2.GaussianBlur(img, (5,5), 0)
_, th_otsu = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# --------------------------
# Step 5: Visualize All Results
# --------------------------
titles = ['Original', 'Binary', 'Binary Inv', 'Truncate', 'ToZero', 'ToZero Inv',
          'Adaptive Mean', 'Adaptive Gaussian', "Otsu's"]

images = [img, th_binary, th_binary_inv, th_trunc, th_tozero, th_tozero_inv,
          th_adaptive_mean, th_adaptive_gaussian, th_otsu]

plt.figure(figsize=(15,10))
for i in range(len(images)):
    plt.subplot(3,3,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
