# Train YOLOv11n-Seg

```python
from ultralytics import YOLO

model = YOLO("yolov8n-seg.pt")

model.train(
    data="/content/Dengue--2/data.yaml",
    epochs=50,
    imgsz=640,
    batch=8,
    name="puddle_seg_yolov8n",
    device=0
)
```
